````python
import numpy as np
import pandas as pd
from textblob import TextBlob
import plotly.graph_objects as go
import plotly.express as xp
````

## **Modi Dataset**

````python
modi = pd.read_csv("/content/drive/MyDrive/Election Prediction/dataset/modi_reviews.csv")
modi.head()
````

Converting all the data samples to string

````python
modi['Tweet'] = modi['Tweet'].astype(str)
````

Define the function to find the polarity of each tweet

````python
def find_polarity(review):
  return TextBlob(review).sentiment.polarity
````

Finding the polarity of all the tweets

````python
modi['Polarity'] = modi['Tweet'].apply(find_polarity)
modi
````

Based on the polarity, tweets are classified into +ve, -ve or neutral

````python
modi['Label'] = np.where(modi['Polarity']>0, 'positive','negative')
modi['Label'][modi['Polarity']==0]='neutral'
modi
````

## **Comparison Graph**

````python
politicians = ['Modi', 'Rahul']

neg_list = [neg_modi, neg_rahul]
pos_list = [pos_modi, pos_rahul]
neu_list = [neu_modi]

fig = go.Figure(
    data = [
        go.Bar(name='Negative', x=politicians, y=neg_list, marker=dict(color='blue')),
        go.Bar(name='Positive', x=politicians, y=pos_list, marker=dict(color='green')),
        go.Bar(name='Neutral', x=politicians, y=neu_list, marker=dict(color='yellow'))
    ]
)

fig.show()
````
